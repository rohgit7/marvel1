```python
import pandas as pd

column_data = soup.find_all('h2')
tasks = [item.text.strip() for item in column_data]

level_titles = ["Level 1", "Level 2", "Level 3","Level 4"]
tasks_by_level = {level: [] for level in level_titles}
current_level = 0

level_1 = tasks[:19]  
level_2 = tasks[19:29]  
level_3 = tasks[29:39] 
level_4 = tasks[39:]

tasks_by_level = {
    "Level 1": level_1,
    "Level 2": level_2,
    "Level 3": level_3,
    "Level 4": level_4
}

# Convert to DataFrame
df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in tasks_by_level.items()]))
print(df)
```
